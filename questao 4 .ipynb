### QUESTAO 2 ###



```python
Caminho2 ='/content/Automoveis.xlsx'
df2=pd.read_excel(Caminho2)
df2.head( )
```

```python
df2.corr( )
```

```python
modelo2=ols('Produção ~ Tempo',data=df2).fit()
print(modelo2.summary())
```

```python
x=df2[['Produção']]
y=df2['Tempo']

backselect2 = step_reg.backward_regression(x, y, 0.05)
backselect2
```

```python
Tempo_estimado = 18

novos_dados = pd.DataFrame({'Tempo': [Tempo_estimado]})

previsao_Producao = modelo2.predict(novos_dados)

print("Previsão de Producao:", previsao_Producao[0])
```

ANALISE:



*   Esta análise mostra uma forte relação linear entre o tempo e a produção, com um coeficiente de determinação (R-quadrado) alto, indicando uma boa adequação do modelo aos dados.

```python
!pip install stepwise-regression

Caminho_3='/content/Multifator.xlsx'
df3=pd.read_excel(Caminho_3)
df3.head( )
```

```python
df3.corr( )
```

```python
modelo=ols('CIF ~ HMOD+HM',data=df3).fit()
print(modelo.summary())
```

```python
modelo_HM =ols('CIF ~ HM',data=df3).fit()
print(modelo_HM.summary())
```

```python
modelo_HMOD=ols('CIF ~ HMOD',data=df3).fit()
print(modelo_HMOD.summary())
```

```python
from stepwise_regression import step_reg
x=df3[['HM','HMOD']]
y=df3['CIF']

backselect3 = step_reg.backward_regression(x,y,0.05)
print(backselect3)
```

ANALISE:


* A variável HM tem um impacto mais significativo no CIF do que HMOD, conforme indicado pelos valores de R-quadrado.


* A regressão múltipla que inclui ambas as variáveis (HMOD e HM) apresenta o melhor ajuste geral, conforme indicado pelo R-quadrado.
